```python
for i in range(len(payroll.drop_duplicates())):
    row = payroll.drop_duplicates().iloc[i]

    # Calculate the salary
    base_salary = row['Base Salary']
    bonus = row['Bonus']
    deductions = row['Deductions']
    tax_rate = row['Tax Rate']
    calculated_salary = (base_salary + bonus - deductions) * ((100 - tax_rate) / 100)

    # Establish relationship employee --> departments
    graph.run("MATCH (p:EMPLOYEES {`First Name`: '" + row['First Name'] + "', `Last Name`: '" + row['Last Name'] + "', `Manager`: '" + row['Manager'] + "', `Department ID`: " + str(row['Department ID']) + ", `Salary ID`: " + str(row['Salary ID']) + ", `Tax ID`: " + str(row['Tax ID']) + "}), (d:DEPARTMENTS {`Department Name`: '" + str(row['Department Name']) + "', `Department ID`: " + str(row['Department ID']) + ", `Location`: '" + row['Location'] + "', `Budget`: " + str(row['Budget']) + "}) CREATE (p)-[:WORKS_AT]->(d)")

    # Establish relationship employee --> salaries 
    graph.run("MATCH (p:EMPLOYEES {`First Name`: '" + row['First Name'] + "', `Last Name`: '" + row['Last Name'] + "', `Manager`: '" + row['Manager'] + "', `Department ID`: " + str(row['Department ID']) + ", `Salary ID`: " + str(row['Salary ID']) + ", `Tax ID`: " + str(row['Tax ID']) + "}), (s:SALARIES {`Base Salary`: " + str(row['Base Salary']) + ", `Salary ID`: " + str(row['Salary ID']) + ", `Bonus`: " + str(row['Bonus']) + ", `Deductions`: " + str(row['Deductions']) + "}) CREATE (p)-[:HAS_BASE_SALARY]->(s)")

    # Create CALCULATED_SALARY node
    graph.run("CREATE (:CALCULATED_SALARY {`Employee ID`: " + str(row['id']) + ", `Calculated Salary`: " + str(calculated_salary) + "})")

    # Establish relationship employee --> calculated_salary
    graph.run("MATCH (p:EMPLOYEES {`First Name`: '" + row['First Name'] + "', `Last Name`: '" + row['Last Name'] + "', `Manager`: '" + row['Manager'] + "', `Department ID`: " + str(row['Department ID']) + ", `Salary ID`: " + str(row['Salary ID']) + ", `Tax ID`: " + str(row['Tax ID']) + "}), (c:CALCULATED_SALARY {`Employee ID`: " + str(row['id']) + ", `Calculated Salary`: " + str(calculated_salary) + "}) CREATE (p)-[:HAS_CALCULATED_SALARY]->(c)")

    # Establish relationship employee --> taxes
    graph.run("MATCH (p:EMPLOYEES {`First Name`: '" + row['First Name'] + "', `Last Name`: '" + row['Last Name'] + "', `Manager`: '" + row['Manager'] + "', `Department ID`: " + str(row['Department ID']) + ", `Salary ID`: " + str(row['Salary ID']) + ", `Tax ID`: " + str(row['Tax ID']) + "}), (t:TAXES {`Tax Rate`: " + str(row['Tax Rate']) + ", `Tax ID`: " + str(row['Tax ID']) + ", `Exemptions`: " + str(row['Exemptions']) + ", `Withholdings`: " + str(row['Withholdings']) + "}) CREATE (p)-[:HAS_TAXES]->(t)")
```
